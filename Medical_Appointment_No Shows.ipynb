### Cleaning the dataset

```python
# standardize column names
```

```python
df = df.rename(columns={
    "Hipertension": "Hypertension",
    "Handcap": "Handicap",
    "No-show": "Attended"
    
})
# by renaming column names
```

```python
df.head()
```

```python
df.dtypes
```

```python
# Now i change the datatype of the datasets into their correct format
```

```python
# datatype of PatientId(Float) -> converting to int first than sting datatype using apply method
# with a lambda function
df['PatientId'] = df['PatientId'].apply(lambda x: str(int(x)))
```

```python
# Converting the ScheduledDay & AppointmentDay to their correct datetime format
df['ScheduledDay']   = pd.to_datetime(df['ScheduledDay'], errors='coerce', utc=True).dt.tz_convert(None)
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay'], errors='coerce', utc=True).dt.tz_convert(None)

```

```python
df.head()
```

```python
# now i can modify some changes is Neighbourhood so i can get first letter of every word to capital
# by using .title()
# strip() it removes the extra spaces if there

df['Neighbourhood'] = df['Neighbourhood'].str.strip().str.title()
```

```python
# In Gender coloumn we unify the values so they’re consistent ('F' and 'M')
# because sometimes datasets have variations like: ("F", "f", "Female", "female")/("M", "m", "Male", "male")

df['Gender'] = df['Gender'].str.strip().str.upper().map({'FEMALE':'F', 'MALE':'M', 'F':'F', 'M':'M'})
```

### Checking outliers if any

```python
df['Age'].describe()
```

```python
# using IQR method
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
lower_bound
```

```python
upper_bound
```

```python
# Negative ages
negative_ages = df[df['Age'] < 0]
negative_ages

# we got lowerbound -37.5 but Age can't be nagative therefore we can drop the negative age row
```

```python
age_outliers = df[(df['Age'] < lower_bound) | (df['Age'] > upper_bound)]
age_outliers
# these are the outliers
```

```python
# Remove negative ages
df = df[df['Age'] >= 0]
```

```python
negative_ages = df[df['Age'] < 0]
negative_ages
```

```python
# Technically the Age 115 is possible,But in most real-world > 100 age is extremely rare.
# we can remove the outlier age 115
df = df[df['Age'] != 115]
```

```python
age_outliers = df[(df['Age'] < lower_bound) | (df['Age'] > upper_bound)]
age_outliers
```
